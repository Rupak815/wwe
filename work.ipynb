```python
travel.drop(columns=['Age'],inplace=True)
travel.drop(columns=['ProductPitched'],inplace=True)
```

```python
cols = [
    'CustomerID', 'DurationOfPitch', 'NumberOfPersonVisiting',
    'NumberOfFollowups', 'PreferredPropertyStar',
    'NumberOfTrips', 'PitchSatisfactionScore',
    'MonthlyIncome', 'NumberOfChildrenVisiting'
]
travel[cols] = travel[cols].fillna(travel[cols].median())
```

```python
travel.isnull().sum()
```

```python
travel['TypeofContact'].fillna(travel['TypeofContact'].mode()[0], inplace=True)
```

```python
travel.isnull().sum()
```

```python
type_contact = pd.get_dummies(travel['TypeofContact'], drop_first=True)
occupation = pd.get_dummies(travel['Occupation'], drop_first=True)
gender = pd.get_dummies(travel['Gender'], drop_first=True)
marital = pd.get_dummies(travel['MaritalStatus'], drop_first=True)
designation = pd.get_dummies(travel['Designation'], drop_first=True)
```

```python
travel.drop(['TypeofContact','Occupation','Gender','MaritalStatus','Designation'],axis=1,inplace=True)
```

```python
travel = pd.concat([travel, type_contact, occupation, gender, marital, designation],axis=1)
```

```python
travel.dtypes
```

```python
travel.isnull().sum()
```

```python
travel
```

```python
x=travel.drop('ProdTaken',axis=1)
y=travel['ProdTaken']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
```

```python
tree_classifier.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled=True)
```

```python
prediction = tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,prediction)
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges")
```

```python
print(classification_report(y_test,prediction))
```
